api_key is the specific key given by the api to access the data.
api_prefix is the start of the link which will be appended to give the full link for the api_end list to give us each leagues matches. Headers is the authentication token. 

```python
api_key = 'd12c14a926214647aeb3b15fd962a0ca'
api_prefix = 'https://api.football-data.org/v4/competitions/'
headers = { 'X-Auth-Token': api_key }
api_end = ['PL/matches/?season=2021', 'BL1/matches/?season=2021', 'FL1/matches/?season=2021', 'SA/matches/?season=2021', 'PD/matches/?season=2021' ]
```

```python
def getMatches():
    #create a list of responses
    responses = []
    #loop through the api_end list and create a linked by adding each api end to the prefix to create a link to the data 
    for i in range(len(api_end)):
        responses.append(requests.get(api_prefix + api_end[i], headers=headers))
        #create a list of dataframes for each league where we will collect their matches 
        DATAFRAMES = []
    for i in range(len(responses)):
        DATAFRAMES.append(pd.DataFrame.from_dict(responses[i].json()["matches"]))
        #concat all 5 leagues into into one dataframe
    Top5Leagues = pd.concat([DATAFRAMES[0],DATAFRAMES[1],DATAFRAMES[2],DATAFRAMES[3], DATAFRAMES[4]])
    #number all games
    Top5Leagues.index = range(len(Top5Leagues))
    #serialise the data using pickle
    Top5Leagues.to_pickle('Top5Leagues.pkl')
    #return the full df
    return Top5Leagues

getMatches()
```
